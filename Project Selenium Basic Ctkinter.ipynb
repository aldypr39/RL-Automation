```python
import tkinter as tk
import customtkinter as ctk
from tkinter import filedialog
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Fungsi untuk memilih file Excel
def select_file():
    global file_path
    file_path = filedialog.askopenfilename(title="Pilih File Excel", filetypes=[("Excel files", "*.xlsx")])
    if file_path:
        # Mengambil hanya nama file dari path lengkap
        file_name = file_path.split("/")[-1]
        file_label.configure(text=f"File Terpilih: {file_name}")
    else:
        file_label.configure(text="Tidak ada file yang dipilih")


# Fungsi untuk memulai proses Selenium
def start_process():
    global file_path, month_choice, rl_choice
    
    # Baca data dari file Excel
    df = pd.read_excel(file_path, header=2)

    # Buka driver Chrome
    driver = webdriver.Chrome()
    driver.get("https://akun-yankes.kemkes.go.id/beranda")
    driver.maximize_window()

    driver.implicitly_wait(15)

    # Login
    driver.find_element(By.ID, "floatingInput").send_keys("rsudulinprovkalsel@gmail.com")
    driver.find_element(By.ID, "floatingPassword").send_keys("Banjarmasin123@")
    time.sleep(2)
    driver.find_element(By.CLASS_NAME, "btn-outline-success").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//*[@id='root']/div/div[2]/a").click()
    driver.implicitly_wait(20)

    # --- LOGIKA BARU BERDASARKAN PILIHAN RL ---
    selected_rl = rl_choice.get()

    if selected_rl == "RL 4.1":
        # Cek dan klik RL.4
        try:
            rl4_element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[@role='button' and text()='RL.4']"))
            )
            driver.execute_script("arguments[0].scrollIntoView(true);", rl4_element)  
            driver.execute_script("arguments[0].click();", rl4_element)
            time.sleep(2)

            dropdown_item = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'RL 4.1 Morbiditas Pasien Rawat Inap')]"))
            )
            dropdown_item.click()
            print("RL 4.1 berhasil diklik")

        except Exception as e:
            print(f"Terjadi masalah: {e}")
            print("Mencoba klik RL.4 lagi...")
            time.sleep(2)
            
            rl4_element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[@role='button' and text()='RL.4']"))
            )
            driver.execute_script("arguments[0].click();", rl4_element)

            dropdown_item = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'RL 4.1 Morbiditas Pasien Rawat Inap')]"))
            )
            dropdown_item.click()
            print("RL 4.1 berhasil diklik (percobaan kedua)")

    elif selected_rl == "RL 5.1":
        # Cek dan klik RL.5 (Anda perlu menyesuaikan XPath di sini)
        try:
            # TODO: Ganti XPath ini untuk RL.5
            rl5_element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[@role='button' and text()='RL.5']"))
            )
            driver.execute_script("arguments[0].scrollIntoView(true);", rl5_element)  
            driver.execute_script("arguments[0].click();", rl5_element)
            time.sleep(2)

            # TODO: Ganti XPath ini untuk RL 5.1
            dropdown_item = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'RL 5.1 Mobiditas Pasien Rawat Jalan')]")) # Contoh XPath
            )
            dropdown_item.click()
            print("RL 5.1 berhasil diklik")

        except Exception as e:
            print(f"Terjadi masalah: {e}")
            print("Mencoba klik RL.5 lagi...")
            time.sleep(2)
            
            # TODO: Ganti XPath ini untuk RL.5
            rl5_element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[@role='button' and text()='RL.5']"))
            )
            driver.execute_script("arguments[0].click();", rl5_element)

            # TODO: Ganti XPath ini untuk RL 5.1
            dropdown_item = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'RL 5.1 Mobiditas Pasien Rawat Jalan')]")) # Contoh XPath
            )
            dropdown_item.click()
            print("RL 5.1 berhasil diklik (percobaan kedua)")
    
    # --- AKHIR LOGIKA BARU ---

    button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//*[@id='root']/div/div/div[1]/div/a"))
    )
    button.click()

    # Convert bulan pilihan ke dalam XPath yang sesuai
    month_mapping = {
        "January": 2, "February": 3, "March": 4, "April": 5, "May": 6, "June": 7,
        "July": 8, "August": 9, "September": 10, "October": 11, "November": 12, "December": 13
    }
    selected_month = month_choice.get()
    month_index = month_mapping[selected_month]

    # Proses input data dari Excel
    for i, row in df.iterrows():
        icd_data = row[2]

        input_element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//input[@name='caripenyakit']"))
        )
        driver.execute_script("arguments[0].scrollIntoView(true);", input_element)
        driver.execute_script("arguments[0].focus();", input_element)
        input_element.clear()  
        input_element.send_keys(icd_data)
        input_element.send_keys(Keys.RETURN)

        add_button = driver.find_element(By.XPATH, "//*[@id='root']/div/div[2]/div/div/div/div[2]/table/tbody/tr[1]/td[4]/button")
        add_button.click()
        time.sleep(2)

        try:
            month_dropdown_xpath = f"//*[@id='bulan']/option[{month_index}]"
            month_dropdown_element = WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH, month_dropdown_xpath))
            )
            month_dropdown_element.click()
            time.sleep(2)
        except Exception as e:
            print(f"Terjadi masalah dalam memilih bulan: {e}")
            return
        
        time.sleep(2)

        # Mengisi kolom umur 1-25
        for j in range(25):
            male_data = row.iloc[4 + 2 * j]
            female_data = row.iloc[5 + 2 * j]
            male_xpath = f"//*[@id='root']/div/div[2]/div[2]/div/div/form/div[2]/div[2]/table/tbody/tr[{j + 1}]/td[3]/input"
            female_xpath = f"//*[@id='root']/div/div[2]/div[2]/div/div/form/div[2]/div[2]/table/tbody/tr[{j + 1}]/td[4]/input"
            
            male_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, male_xpath))
            )
            driver.execute_script("arguments[0].scrollIntoView(true);", male_element)
            male_element.clear()
            male_element.send_keys(str(male_data))

            female_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, female_xpath))
            )
            driver.execute_script("arguments[0].scrollIntoView(true);", female_element)
            female_element.clear()
            female_element.send_keys(str(female_data))

        # Mengisi kolom umur >65 (kelompok umur ke-26)
        male_last_data = row.iloc[57]
        female_last_data = row.iloc[58]
        male_last_xpath = "//*[@id='root']/div/div[2]/div[2]/div/div/form/div[2]/div[2]/table/tbody/tr[26]/td[3]/input"
        female_last_xpath = "//*[@id='root']/div/div[2]/div[2]/div/div/form/div[2]/div[2]/table/tbody/tr[26]/td[4]/input"

        male_last_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, male_last_xpath))
        )
        driver.execute_script("arguments[0].scrollIntoView(true);", male_last_element)
        male_last_element.clear()
        male_last_element.send_keys(str(male_last_data))

        female_last_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, female_last_xpath))
        )
        driver.execute_script("arguments[0].scrollIntoView(true);", female_last_element)
        female_last_element.clear()
        female_last_element.send_keys(str(female_last_data))

        time.sleep(2)

        save_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "//*[@id='root']/div/div[2]/div[2]/div/div/form/div[3]/button"))
        )
        save_button.click()

        time.sleep(2)

        back_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//*[@id='root']/div/div/div[1]/div/a"))
        )
        back_button.click()

# --- GUI Setup ---
ctk.set_appearance_mode("System")
ctk.set_default_color_theme("blue")

root = ctk.CTk()
root.title("Automasi Input Data")
root.geometry("600x450")

frame = ctk.CTkFrame(master=root)
frame.pack(pady=20, padx=60, fill="both", expand=True)

# Tombol Pilih File
file_button = ctk.CTkButton(master=frame, text="Pilih File Excel", command=select_file)
file_button.pack(pady=12, padx=10)

# Label untuk menampilkan file yang dipilih
file_label = ctk.CTkLabel(master=frame, text="Tidak ada file yang dipilih", wraplength=400)
file_label.pack(pady=12, padx=10)

# Dropdown untuk memilih RL
rl_label = ctk.CTkLabel(master=frame, text="Pilih Jenis RL")
rl_label.pack(pady=(10, 0))

rl_choice = ctk.StringVar(value="RL 4.1")
rl_dropdown = ctk.CTkComboBox(master=frame, values=["RL 4.1", "RL 5.1"], variable=rl_choice)
rl_dropdown.pack(pady=12, padx=10)

# Dropdown untuk memilih bulan
month_label = ctk.CTkLabel(master=frame, text="Pilih Bulan")
month_label.pack(pady=(10, 0))

month_choice = ctk.StringVar(value="January")
month_dropdown = ctk.CTkComboBox(
    master=frame,
    values=["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"],
    variable=month_choice
)
month_dropdown.pack(pady=12, padx=10)

# Tombol untuk memulai proses
start_button = ctk.CTkButton(master=frame, text="Mulai Proses", command=start_process)
start_button.pack(pady=20, padx=10)

root.mainloop()

```
